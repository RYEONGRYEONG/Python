```python
# 외부 라이브러리 작업
    # 가져오기, 연산자 오버로딩 및 외부 라이브러리 세계로의 모험을 위한 생존 팁

"""
이 자습서에서는 Python의 가져오기에 대해 배우고 익숙하지 않은 라이브러리(및 라이브러리가 반환하는 개체) 작업에 대한 몇 가지 팁을 얻고 연산자 오버로딩에 대해 자세히 알아봅니다.¶
"""
""""""
```

```python
# 임포트
"""
지금까지 우리는 언어에 내장된 유형과 함수에 대해 이야기했습니다.

그러나 Python의 가장 좋은 점 중 하나(특히 데이터 과학자인 경우)는 Python용으로 작성된 수많은 고품질 사용자 지정 라이브러리입니다.

이러한 라이브러리 중 일부는 "표준 라이브러리"에 있습니다.
즉, Python을 실행하는 모든 위치에서 찾을 수 있습니다.
다른 라이브러리는 Python과 함께 제공되지 않는 경우에도 쉽게 추가할 수 있습니다.

어느 쪽이든 가져오기를 통해 이 코드에 액세스합니다.

표준 라이브러리에서 math를 가져오는 것으로 예제를 시작하겠습니다.
"""

import math

print("It's math! It has type {}".format(type(math)))
```

```python
"""
math는 모듈입니다.
모듈은 다른 사람이 정의한 변수 모음(원하는 경우 네임스페이스)입니다.
내장 함수 dir()을 사용하여 math에서 모든 이름을 볼 수 있습니다.
"""

print(dir(math))
```

```python
"""
도트 구문을 사용하여 이러한 변수에 액세스할 수 있습니다.
그들 중 일부는 math.pi와 같은 간단한 값을 참조합니다.
"""

print("pi to 4 significant digits = {:.4}".format(math.pi))
```

```python
"""
그러나 모듈에서 찾을 수 있는 대부분은 math.log와 같은 함수입니다.
"""

math.log(32, 2) //밑이 2고 진수가 32, 2 ** 5
```

```python
"""
물론 math.log가 무엇을 하는지 모른다면 help()를 호출할 수 있습니다.
"""

help(math.log)
```

```python
00"""
모듈 자체에서 help()를 호출할 수도 있습니다.
이렇게 하면 모듈의 모든 기능과 값에 대한 결합된 문서가 제공됩니다(모듈에 대한 높은 수준의 설명도 포함).
전체 수학 도움말 페이지를 보려면 "출력" 버튼을 클릭하십시오.
"""

help(math)
```

```python
# 기타 가져오기 구문
"""
우리가 수학에서 함수를 자주 사용한다는 것을 안다면 타이핑을 줄이기 위해 더 짧은 별칭으로 함수를 가져올 수 있습니다(이 경우 "math"는 이미 꽤 짧습니다).
"""

import math as mt 
mt.pi
```

```python
"""
    Pandas, Numpy, Tensorflow 또는 Matplotlib와 같은 인기 있는 특정 라이브러리에서 이 작업을 수행하는 코드를 본 적이 있을 것입니다.
    예를 들어 numpy를 np로 가져오고 pandas를 pd로 가져오는 것이 일반적인 규칙입니다.

as는 단순히 가져온 모듈의 이름을 바꿉니다. 다음과 같은 작업을 수행하는 것과 같습니다.
"""

import math
mt = math # 이 방식보다는 
```

```python
"""
수학 모듈의 모든 변수를 자체적으로 참조할 수 있다면 좋지 않을까요?
즉, math.pi 또는 mt.pi 대신 pi를 참조할 수 있다면?
좋은 소식: 우리는 그렇게 할 수 있습니다.
"""

from math import * # math로부터 전체를 다 가져오겠다.
print(pi, log(32, 2))
```

```python
"""
import *는 모듈의 모든 변수에 직접 액세스할 수 있도록 합니다(점 접두사 없이).

나쁜 소식: 일부 순수주의자들은 당신이 이렇게 하는 것에 대해 투덜거릴 수 있습니다.

더 나쁘다 : 그들은 일종의 요점이 있습니다.
"""

from math import *
from numpy import *
print(pi, log(32, 2)) # 오류남! math에서도 log가 있고 numpy에도 log가 있기때문
```

```python
"""
무슨 일이 일어난?
이전에 작동했습니다!

이러한 종류의 "별표 가져오기"는 때때로 이상하고 디버그하기 어려운 상황으로 이어질 수 있습니다.

이 경우 문제는 math 모듈과 numpy 모듈 모두 log라는 함수가 있지만 의미 체계가 다르다는 것입니다.
numpy second에서 가져오기 때문에 해당 로그는 math에서 가져온 로그 변수를 덮어씁니다(또는 "그림자").

적절한 절충안은 각 모듈에서 필요한 특정 항목만 가져오는 것입니다.
"""

from math import log, pi
from numpy import asarray
```
